# Number of Fires by Neighborhood

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint

fires = pd.read_table("Fire_Data.csv", sep=",")
neighborhoods = fires.groupby("neighborhood")
neighborhood_dict = neighborhoods.size().sort_values().to_dict()

#pprint(neighborhood_dict)
print("neighborhood with least fires:", min(neighborhood_dict, key=neighborhood_dict.get))
print("neighborhood with most fires:", max(neighborhood_dict, key=neighborhood_dict.get))

keys = list(neighborhood_dict.keys())
vals = list(neighborhood_dict.values())

plt.clf()
plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(neighborhood_dict)), vals, align="center", width=0.5)
plt.xticks(range(len(neighborhood_dict)), keys, rotation="vertical")
plt.show()
```

# Types of Fires by Neighborhood

```python
fire_types = fires.groupby("type_description")
fire_types_dict = fire_types.size().sort_values().to_dict()

keys = list(fire_types_dict.keys())
vals = list(fire_types_dict.values())

plt.clf()
plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(fire_types_dict)), vals, align="center", width=0.5)
plt.xticks(range(len(fire_types_dict)), keys, rotation="vertical")
plt.show()
plt.close()

fire_dict = dict()
for fire, frame in fire_types:
    fire_group = frame.groupby("neighborhood")
    fire_dict[fire] = fire_group.size().sort_values().to_dict()

print("neighborhood with least building fires:", min(fire_dict["Building fire"], key=fire_dict["Building fire"].get))
print("neighborhood with most building fires:", max(fire_dict["Building fire"], key=fire_dict["Building fire"].get))

keys = list(fire_dict["Building fire"].keys())
vals = list(fire_dict["Building fire"].values())

plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(fire_dict["Building fire"])), vals, align="center", width=0.5)
plt.xticks(range(len(fire_dict["Building fire"])), keys, rotation="vertical")
plt.show()
plt.close()

keys = list(fire_dict["Cooking fire, confined to container"].keys())
vals = list(fire_dict["Cooking fire, confined to container"].values())

plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(fire_dict["Cooking fire, confined to container"])), vals, align="center", width=0.5)
plt.xticks(range(len(fire_dict["Cooking fire, confined to container"])), keys, rotation="vertical")
plt.show()
plt.close()
```
