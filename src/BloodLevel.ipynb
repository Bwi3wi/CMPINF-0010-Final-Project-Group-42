```python

# This code set sorts the blood lead level from highest to lowest by school district chronologically between the years 2015-2020



import pandas as pd


order = pd.read_csv('Bloodlevel.csv')


order_2015 = order[['SchoolDistrict', 'percentEBLL2015']]


sorted_order_2015 = order_2015.sort_values(by='percentEBLL2015', ascending=False)


print(sorted_order_2015)
print()

order_2016=order[['SchoolDistrict','percentEBLL2016']]
sorted_order_2016=order_2016.sort_values(by='percentEBLL2016', ascending=False)

print(sorted_order_2016)
print()

order_2017=order[['SchoolDistrict','percentEBLL2017']]
sorted_order_2017=order_2017.sort_values(by='percentEBLL2017', ascending=False)
print(sorted_order_2017)
print()


order_2018=order[['SchoolDistrict','percentEBLL2018']]
sorted_order_2018=order_2018.sort_values(by='percentEBLL2018', ascending=False)
print(sorted_order_2018)

print()


order_2019=order[['SchoolDistrict','percentEBLL2019']]
sorted_order_2019=order_2019.sort_values(by='percentEBLL2019', ascending=False)
print(sorted_order_2019)
print()


order_2020=order[['SchoolDistrict','percent2020']]
sorted_order_2020=order_2020.sort_values(by='percent2020', ascending=False)
print(sorted_order_2020)
print()
```

```python
#graph

import pandas as pd
import matplotlib.pyplot as plt
graph = pd.read_csv('Bloodlevel.csv')
graph_sorted = graph.sort_values(by='percentEBLL2019')
graph_sorted.plot(x='SchoolDistrict', y='percentEBLL2019', kind='barh')
plt.xlabel('Percent of children with elevated blood lead levels in 2019')
plt.ylabel('School District')
plt.title('Percent of Children with Elevated Blood Lead Levels by School District in 2019')
plt.show()
```

```python

#randomly choses 15 school from the datset and shows the schools with the highest blood lead level
# are able to change the year hence changing the graph


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


high = pd.read_csv('Bloodlevel.csv')
schools = np.random.choice(high['SchoolDistrict'].unique(), size=15, replace=False)
high_filtered = high[high['SchoolDistrict'].isin(schools)]
high_sorted = high_filtered.sort_values(by=['percentEBLL2015'], ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(high_sorted['SchoolDistrict'], high_sorted['percentEBLL2016'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('SchoolDistrict')
plt.ylabel('Blood Lead Level')
plt.title('Blood Lead Levels by School')
plt.show()



```
