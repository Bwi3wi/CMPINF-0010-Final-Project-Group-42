# Neighborhood Safety in Pittsburgh

## Introduction and the Metric

This project attempts to determine the "best" neighborhood in Pittsburgh based on whether or not someone would like to live there over another neighborhood. However, there are many metrics to consider when deciding on a new place to live, so in order to provide a clear measure of how livable a neighborhood is it was decided to determine the "best" neighborhood in Pittsburgh based on a metric of how safe it is. It is believed that the metric of "safety" should be the deciding factor of the "best" neighborhood in Pittsburgh because overall it is the safety of the public that matters most, as no other metric matters if individuals living in an area are not safe.

To determine the "best" neighborhood in Pittsburgh based on how safe it is, 3 submetrics with corresponding datasets were analyzed. The first submetric being considered to determine safety is the level of lead found in the blood of children across various school districts in the Pittsburgh and Allegheny area. With Pittsburgh being an older city with lead piping systems, the amount of lead found in water across the city is a deep concern affecting the health of the public, so it is believed that neighborhoods with school districts that have a lesser chance of children containing harmful levels of lead in their blood will be safer overall. The second submetric being considered to determine safety is how many arrests are made in a neighborhood compared to other neighborhoods. The number of arrests in a neighborhood is related deeply to the amount of crime in that neighborhood, so it is assumed that a neighborhood with a lower number of arrests is safer than one with more arrests. The final submetric used to consider safety is how many fire incidents occur in a neighborhood. It is assumed that a neighborhood with less fires is overall safer than a neighborhood with more, as more fire incidents indicate a higher liklihood of someone being harmed in the process.

## Submetric: Blood Lead Level

The dataset being analyzed to determine safety based on the amount of lead found in blood can be found in the file "Bloodlevel.csv". This dataset reports the percentage of children found in each school district to have blood lead levels that could be considered harmful. The code analysis can be found below:

```python

# This code set sorts the blood lead level from highest to lowest by school district chronologically between the years 2015-2020



import pandas as pd


order = pd.read_csv('Bloodlevel.csv')


order_2015 = order[['SchoolDistrict', 'percentEBLL2015']]


sorted_order_2015 = order_2015.sort_values(by='percentEBLL2015', ascending=False)


print(sorted_order_2015)
print()

order_2016=order[['SchoolDistrict','percentEBLL2016']]
sorted_order_2016=order_2016.sort_values(by='percentEBLL2016', ascending=False)

print(sorted_order_2016)
print()

order_2017=order[['SchoolDistrict','percentEBLL2017']]
sorted_order_2017=order_2017.sort_values(by='percentEBLL2017', ascending=False)
print(sorted_order_2017)
print()


order_2018=order[['SchoolDistrict','percentEBLL2018']]
sorted_order_2018=order_2018.sort_values(by='percentEBLL2018', ascending=False)
print(sorted_order_2018)

print()


order_2019=order[['SchoolDistrict','percentEBLL2019']]
sorted_order_2019=order_2019.sort_values(by='percentEBLL2019', ascending=False)
print(sorted_order_2019)
print()


order_2020=order[['SchoolDistrict','percent2020']]
sorted_order_2020=order_2020.sort_values(by='percent2020', ascending=False)
print(sorted_order_2020)
print()
```

Chosing which neighborhood area is safest in terms of blood lead level found between 2015-2020

```python
# determine safest school district across all years

import pandas as pd
safe = pd.read_csv('Bloodlevel.csv')
indices = ['percentEBLL2015','percentEBLL2016','percentEBLL2017','percentEBLL2018','percentEBLL2019','percent2020']
avg_lead_levels = safe.groupby('SchoolDistrict')[indices].mean()
safest_school = avg_lead_levels.mean(axis=1).idxmin()
print(f"The safest school based on blood lead level is {safest_school}")
```

## Submetric: Fire Incidents by Neighborhood

The dataset being analyzed to determine safety based on the fire incidents in a neighborhood can be found in the file "Fire_Data.csv". This dataset reports the fire incidents that occur in the city of Pittsburgh as well as what neighborhood they occur in and the kind of fire it is. The code analysis can be found below:

```python
# Report the neighborhood with the least and most fires from the dataset

import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint

# read dataset and group by neighborhood
fires = pd.read_table("Fire_Data.csv", sep=",")
neighborhoods = fires.groupby("neighborhood")

# Get size of each group to determine neighborhood with least and most fires
neighborhood_dict = neighborhoods.size().sort_values().to_dict()

# print results
print("neighborhood with least fires:", min(neighborhood_dict, key=neighborhood_dict.get))
print("neighborhood with most fires:", max(neighborhood_dict, key=neighborhood_dict.get))


# graph the results of the entire dataset to visualize the data
keys = list(neighborhood_dict.keys())
vals = list(neighborhood_dict.values())
plt.clf()
plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(neighborhood_dict)), vals, align="center", width=0.5)
plt.xticks(range(len(neighborhood_dict)), keys, rotation="vertical")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Fire Incidents")
plt.title("Number of Fires by Neighborhood")
plt.show()
```

### Types of Fires by Neighborhood

Although it is important to consider all possibilities of a fire incident within a neighborhood, some fire types may not be as worrisome to residents as others. As such, the below code segment determines the most common fire types witihn the city of Pittsburgh as well as how they compare to other neighborhoods in these fire types. The most common fire types were found to be building fires and cooking fires, which are important to consider as these could massively affect the safety of individuals if they live in the buildings that the fires occur in.

```python
# Determine most common fire types and report how often they occur in each neighborhood

# group the fire types and sort by number of occurrences
fire_types = fires.groupby("type_description")
fire_types_dict = fire_types.size().sort_values().to_dict()

# graph fire types by occurrences
keys = list(fire_types_dict.keys())
vals = list(fire_types_dict.values())
plt.clf()
plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(fire_types_dict)), vals, align="center", width=0.5)
plt.xticks(range(len(fire_types_dict)), keys, rotation="vertical")
plt.xlabel("Fire Type")
plt.ylabel("Occurrences")
plt.title("Fire Types by Occurrences")
plt.show()
plt.close()

# sort fire types dictionary by number of fire occurrences
fire_dict = dict()
for fire, frame in fire_types:
    fire_group = frame.groupby("neighborhood")
    fire_dict[fire] = fire_group.size().sort_values().to_dict()

# print neighborhoods with most and least building fires
print("neighborhood with least building fires:", min(fire_dict["Building fire"], key=fire_dict["Building fire"].get))
print("neighborhood with most building fires:", max(fire_dict["Building fire"], key=fire_dict["Building fire"].get))

# graph neighborhoods by building fire occurrences
keys = list(fire_dict["Building fire"].keys())
vals = list(fire_dict["Building fire"].values())
plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(fire_dict["Building fire"])), vals, align="center", width=0.5)
plt.xticks(range(len(fire_dict["Building fire"])), keys, rotation="vertical")
plt.xlabel("Neighborhood")
plt.ylabel("Fire Incidents")
plt.title("Number of Building Fires by Neighborhood")
plt.show()
plt.close()

# print neighborhoods with most and least cooking fires
print("neighborhood with least cooking fires:", min(fire_dict["Cooking fire, confined to container"], key=fire_dict["Cooking fire, confined to container"].get))
print("neighborhood with most cooking fires:", max(fire_dict["Cooking fire, confined to container"], key=fire_dict["Cooking fire, confined to container"].get))

# graph neighborhoods by cooking fire occurrences
keys = list(fire_dict["Cooking fire, confined to container"].keys())
vals = list(fire_dict["Cooking fire, confined to container"].values())
plt.rcParams["figure.figsize"] = (20, 8)
plt.bar(range(len(fire_dict["Cooking fire, confined to container"])), vals, align="center", width=0.5)
plt.xticks(range(len(fire_dict["Cooking fire, confined to container"])), keys, rotation="vertical")
plt.xlabel("Neighborhood")
plt.ylabel("Fire Incidents")
plt.title("Number of Cooking Fires by Neighborhood")
plt.show()
plt.close()
```

### Submetric Conclusion

Overall for this submetric, it appears based on the data visualizations that Ridgemont is the safest neighborhood to live in based on fire incidents. Although Ridgemont does not have the least number of fire occurrences, it is important to consider that it has no occurrences of the most common fire types in building and cooking fire, while the other neighborhoods with less total fires do.
